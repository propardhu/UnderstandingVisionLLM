```python
import torch
from PIL import Image
from transformers import AutoProcessor, AutoModelForVision2Seq

# âœ… Detect device (Use MPS for Mac)
DEVICE = "mps" if torch.backends.mps.is_available() else "cpu"
print(f"Using device: {DEVICE}")

# âœ… Load a single image from local disk (Change file path as needed)
image_path = "./image.jpg"  # Update with your actual path
image = Image.open(image_path).convert("RGB")

# âœ… Initialize processor and model
processor = AutoProcessor.from_pretrained("llava-hf/llava-1.5-7b-hf")
model = AutoModelForVision2Seq.from_pretrained(
    "llava-hf/llava-1.5-7b-hf",
    torch_dtype=torch.float16,  # Use float16 for efficiency on Mac
    attn_implementation="eager",  # Required for Transformers 5.0+
    output_attentions=True,  # Enable attention extraction
    return_dict=True,  # Return output as dictionary
    return_dict_in_generate=True
).to(DEVICE)

print("âœ… Model & Processor Loaded!")
```

```python
# âœ… Prepare input prompt
messages = [
    {
        "role": "user",
        "content": [
            {"type": "image"},
            {"type": "text", "text": "What is the main subject here?"}
        ]
    },
]

# âœ… Tokenize text prompt
prompt = processor.apply_chat_template(messages, add_generation_prompt=True)

# âœ… Preprocess image and text
inputs = processor(
    text=prompt,
    images=[image],
    return_tensors="pt"
).to(DEVICE)

print("âœ… Inputs processed, running inference...")

# âœ… Generate output with attention extraction
with torch.no_grad():
    outputs = model.generate(**inputs, max_new_tokens=500, output_attentions=True, return_dict_in_generate=True)

# âœ… Decode the generated response
generated_texts = processor.batch_decode(outputs.sequences, skip_special_tokens=True)
generated_answer = generated_texts[0]
print("\nðŸ“ **Generated Answer:**")
print("â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”")
print(generated_answer)
print("â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”â€”\n")

# âœ… Extract attentions from model output
attentions = outputs.attentions  # Extract all attention layers
print(f"âœ… Extracted {len(attentions)} attention layers.")
```

```python
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
import numpy as np

# âœ… Contrast Stretching Function
def contrast_stretch(attn_map):
    p2, p98 = np.percentile(attn_map, (2, 98))  # Get 2nd and 98th percentiles
    attn_map = np.clip((attn_map - p2) / (p98 - p2), 0, 1)  # Stretch contrast
    return attn_map

# âœ… Gamma Correction Function (Enhances Attention Areas)
def gamma_correction(attn_map, gamma=1.5):
    return np.power(attn_map, gamma)

# âœ… Iterate through each decoder layer's attention
for layer_idx, layer_attn in enumerate(attentions):
    print(f"Processing attention map for Layer {layer_idx + 1}")

    if isinstance(layer_attn, tuple):  
        layer_attn = layer_attn[0]  # âœ… Extract the first element if it's a tuple

    # âœ… Average attention across all heads
    attn_map = layer_attn.mean(dim=1).cpu().detach().numpy().squeeze()

    # âœ… Check if attention is 1D & reshape it
    if attn_map.ndim == 1:  
        sqrt_len = int(np.sqrt(attn_map.shape[0]))  # Try to infer the correct 2D shape
        if sqrt_len * sqrt_len == attn_map.shape[0]:
            attn_map = attn_map.reshape(sqrt_len, sqrt_len)  # Reshape to square if possible
        else:
            print(f"âš ï¸ Skipping Layer {layer_idx + 1}: Non-square attention shape {attn_map.shape}")
            continue  # Skip this layer

    # âœ… Normalize the attention map safely
    attn_map = (attn_map - attn_map.min()) / (attn_map.max() - attn_map.min() + 1e-8)  # Avoid division by zero

    # âœ… Apply Contrast Stretching & Gamma Correction
    attn_map = contrast_stretch(attn_map)
    attn_map = gamma_correction(attn_map, gamma=1.7)  # More contrast

    # âœ… Convert to Torch Tensor and add batch dimension
    attn_map_tensor = torch.tensor(attn_map).unsqueeze(0).unsqueeze(0)  # Shape (1,1,H,W)

    # âœ… Resize attention map to match image dimensions
    resize_transform = transforms.Resize((image.height, image.width), antialias=True)
    attn_map_resized = resize_transform(attn_map_tensor).squeeze().numpy()

    # âœ… Plot the Original Image & Attention Map
    plt.figure(figsize=(12, 6))

    # ðŸ”¹ Original Image
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.axis("off")
    plt.title("Original Image")

    # ðŸ”¹ Attention Heatmap
    plt.subplot(1, 2, 2)
    plt.imshow(image)
    plt.imshow(attn_map_resized, cmap="inferno", alpha=0.5)
    plt.axis("off")
    plt.title(f"Attention Heatmap - Layer {layer_idx+1}")

    # âœ… Save each heatmap separately
    heatmap_filename = f"attention_layer_{layer_idx+1}.png"
    plt.savefig(heatmap_filename, dpi=300)
    print(f"âœ… Saved heatmap: {heatmap_filename}")

plt.show()
```
